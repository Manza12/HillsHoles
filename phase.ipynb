```python
import numpy as np
import scipy.signal as sig
```

```python
f_0 = 3
phi_0 = 0
n = 100
t = np.linspace(0, 1, n)
x = np.sin(2 * np.pi * (f_0 * t + phi_0))
```

```python
def create_sinus(freq, phase):
    return np.sin(2 * np.pi * (f_0 * t + phi_0))

def create_spectrogram(x):
    tau, omega, z = sig.stft(x, nperseg=16, nfft=256)
    spectrogram = 10 * np.log10(np.abs(z)**2 + 0.000001)
    return spectrogram
    
```

```python
spectrogram = create_spectrogram(x)
```

```python
from ipywidgets import interact, interactive, fixed, interact_manual
import ipywidgets as widgets
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
fig, axs = plt.subplots(1, 2)
line, = axs[0].plot(t, x)
im = axs[1].imshow(spectrogram, origin='lower', aspect='auto', vmin=-120, cmap='afmhot')

def plot_sinus(x):
    line.set_data(t, x)

def plot_spectrogram(spectrogram):
    im.set_data(spectrogram)

def plot_sinus_spectrogram(freq, phase):
    print(freq, phase)
    x = create_sinus(freq, phase)
    create_spectrogram(x)
    
    plot_sinus(x)
    plot_spectrogram(spectrogram)

interact(plot_sinus_spectrogram, 
         freq = widgets.FloatSlider(value=5., min=1., max=10., step=1.),
         phase = widgets.FloatSlider(value=0., min=0., max=1.0, step=0.05),
        )
```

```python
def plot_sinus(freq, phase):
    x = create_sinus(freq, phase)
    plt.plot(t, x)


interact(plot_sinus, 
         freq = widgets.FloatSlider(value=5., min=1., max=10., step=1.),
         phase = widgets.FloatSlider(value=0., min=0., max=1.0, step=0.05),
        )
```

```python
def plot_spectrogram(freq, phase):
    x = create_sinus(freq, phase)
    spectrogram = create_spectrogram(x)
    plt.imshow(spectrogram, origin='lower', aspect='auto', vmin=-100, cmap='afmhot')


interact(plot_spectrogram, 
         freq = widgets.FloatSlider(value=5., min=1., max=10., step=1.),
         phase = widgets.FloatSlider(value=0., min=0., max=1.0, step=0.05),
        )
```
